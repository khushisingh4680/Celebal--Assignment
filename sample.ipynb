This function is created to clean numeric value by removing symbols such as ₹,commas,quotes and decimal points

```python
def fix_numbers(x):
    if isinstance(x,str):
        return x.replace('₹', '').replace(',', '').replace('.', '').replace('"', '')
    return x
```

apply() is used to apply the function to a column

```python
df['final_price'] =df['final_price'].apply(fix_numbers)
```

```python
df
```

astype() is used to change the datatype of the column

```python
df['final_price'] = df['final_price'].astype(int)
df['final_price'] = df['final_price']/100
```
